$ OSCILADOR ~ ARMONICO ~ UNIDIMENSIONAL$

Un oscilador armonico es un sistema fisico que se compone por un resorte(muelle) fijo a un extremo, y al otro extremo
una masa m, la cual esta sujeta a moverse con el extremo del resorte.

Es conocido que la forma oscilatoria de un muelle bajo condiciones ideales, éste presenta una cierta caracteristica,
como ejemeplo  la Amplitud de oscilacion permanece constante en el tiempo.

Sin embargo, si el sistema se encuentra bajo un fator de rozamiento, su amplitud decae con el tiempo, anulandose el 
moviemiento para un tiempo relativamente grande, llegando a su origen(estado de reposo); Para mantener la oscilación
es necesario aplicar una fuerza oscilante $F_o cos(w_f t)$, que obligue al resorte seguir en movimiento, esto se conoce como $ oscilación forzada$.

la amplitud de la oscilacion ahora cambiará según como se aplique la fuerza oscilante, y su amplitud maxima se obtendra cuando la frecuaencia de la fuerza oscilante se iguale con la frecuencia propia del oscilador, encontrando el fenomeno de resonancia.

La ecuacion que mide la amplitud del oscilador es de la forma $A = \frac{F_o}{m\sqrt{(w_f ^2 -w_o ^2)^2 +4m^2 \gamma ^2}}$ Donde $ \gamma$ representa la constante de amortiguamiento y $ w_f$ la frecuencia de la fuerza oscilante; Al existir una fuerza de rozamiento, la energía se está transformando continuamente en calor. Puesto que la energía almacenada, en promedio, se conserva, esta energía que se pierde debe ser aportada regularmente por un agente externo, en este caso, la fuerza aplicada sobre la partícula, la potencia con la que se disipa la energía es $p = \frac{F_o \gamma w_f ^2}{m((w_f ^2 -w_o ^2)^2 +4m^2 \gamma ^2)}$, esta ecuación viene de calcular el valor medio de la energía por unidad de tiempo suministrada por la fuerza oscilante $ p = < F_o cos(w_f t) v >$, podemos tomar la expresion de la potencia de una forma mas sencilla $ p = \frac{F_o}{4m\gamma^2}(\frac{1}{1+x^2})$, donde $x = \frac{w_f ^2 - w_o ^2}{2 \gamma w_f}$.

La representación de la potencia P en función de X tiene la forma de una campana(curva). El máximo de la potencia P se obtiene para X=0, o bien, cuando la frecuencia $w_f$ de la fuerza oscilante es igual a la frecuencia $w_0$ natural del oscilador. 

```python
import numpy as np
from numpy import loadtxt
import matplotlib.pyplot as plt
from scipy import linspace
from Distribucion import Distribucion
```

```python

x = linspace(-2,2,100)
# el factor F_0 /4m\gamma^2 se iguala a 1 por simplicidad
p = 1/(1+x**2)
plt.plot(x,p)
plt.xlabel('x')
plt.ylabel('p')
```

Vemos que P tiende rápidamente a cero a medida que X se hace mayor o menor que cero, es decir, a medida que la frecuencia $w_f$ de la fuerza oscilante se hace mayor o menor que la frecuencia $w_0$ propia del oscilador.

Cuando la frecuencia $w_f$ de la fuerza oscilante es igual a la frecuencia $w_0$ natural del oscilador la fuerza oscilante $F_0 cos(w_f t)$  y la velocidad v del oscilador están en fase, el valor medio de la energía por unidad de tiempo P suministrada por la fuerza oscilante es máxima. Esta situación  recibe el nombre de $resonancia$.

La  anchura es otra característica importante de la curva. Se define como el intervalo de frecuencias de la fuerza oscilante para los cuales la potencia P es mayor que la mitad de la máxima. El intervalo de frecuencias $w_f$ alrededor de la frecuencia $w_0$ propia del oscilador está comprendido entre $X=-1$ a $X=+1$, y vale aproximadamente $2 \gamma$.
