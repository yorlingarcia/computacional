```python
import numpy as np
from sympy import *
from scipy import linspace
from matplotlib import pyplot as plt
```

```python
Deltax = 0.1
Xo = 1
Yo = 1
Xmin = 1
Xmax = 1.5
```

```python
h = 0.1
N = (Xmax -Xmin)/h
x = linspace(Xmin,Xmax,int(N))
y = np.zeros(len(x))
y[0] = Yo
```

```python
def funcion(x,y):
    Ecu = 2*x*y
    return(Ecu)
```

```python
for i in range(len(x)-1):
    y[i+1] = y[i] +h*funcion(x[i],y[i])
```

```python
####################
# solucion analitica
Y = np.exp(x**2 -1)
#####################
```

```python
plt.plot(x,y)
plt.plot(x,Y)
plt.legend(('Euler', 'Analitica'),
prop = {'size':10}, loc = 'upper right')
```

```python
h = 0.05
N = (Xmax -Xmin)/h
xx = linspace(Xmin,Xmax,int(N))
z = np.zeros(len(xx))
Yo = 1
Xo = 1
z[0] = Yo
for i in range(len(xx)-1):
    z[i+1] = z[i] + h*funcion(xx[i],z[i])
```

```python
plt.plot(x,y)
plt.plot(xx,z)
plt.plot(x,Y)
plt.legend(('Euler h = 0.1', 'Euler h = 0.05','Analitica'),
prop = {'size':10}, loc = 'upper right')
```
